```python
# 1. Read the data-> identify what's input and output (x and y)
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/EY_batch11_3Mar/refs/heads/main/data.csv')
data = data.iloc[:, :-1]
```

```python
data.head(10)
```

```python
len(set(data['id']))
```

```python
data.info()
```

```python
df = data.set_index('id')
df.head()
```

```python
# identify x and y
y = df['diagnosis']
y.head()
```

```python
x = df.iloc[:, 1:] # all columns except the first one (Diagnosis)
x.head()
```

```python
# unique values in diagnosis

set(df['diagnosis'])
```

```python
#logic = lambda input: 1 if input == 'M' else 0
d_logic = { 'M':1, 'B': 0}
logic = lambda input: d_logic[input]
y = y.map(logic) # map = apply this logic to every data point!
```

```python
y.tail()
```

```python
# 3. Split the data -> train, test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
xtrain.head(2)
```

```python
ytrain.head(2)
```

```python
len(xtrain), len(xtest)
```

```python
# 4. On training data we will try to fit multiple algorithms
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression #Classification algos more than regression :)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain) # inputs and labels!
```

```python
predictions = model.predict(xtest) # only test data, no known results
```

```python
predictions[:5]
```

```python
ytest[:5]
```

```python
# out predictions are on a different scale than diagnosis (known result)
predictions.min(), predictions.max()
```

```python
# let's find a threshold in between

threshold = (predictions.min() + predictions.max())/2.0
threshold
```

```python
predicition_fn = lambda n : 1 if n >= threshold else 0
p = [ predicition_fn(pr) for pr in predictions]
```

```python
p[:5]
```

```python
ytest[:5]
```

```python
# evaluate the predictions against known results

# accuracy = correct guesses / total guesses
# accuracy is between 0 to 1, or 0% to 100%

from sklearn.metrics import accuracy_score
accuracy_score(ytest, p)

```

```python

# lets try next model

model2 = LogisticRegression()
model2.fit(xtrain, ytrain)
p2 = model2.predict(xtest)
#print(p2[:5]) # feel free to uncomment- this shows that predictions are already 0 and 1, no need to threshold
print(accuracy_score(ytest, p2))
```

```python
# joblib and pickle -> these 2 libraries can be used to save the model, and later load it
import joblib
joblib.dump(model2, 'trained_model.joblib')
```
